# Psychopy Reach Angle Data




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Import matplotlib colormaps
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap


%matplotlib inline
%config Inline.figure_format = 'retina'
```

```python
#Import csv and create dataframe for each subject

Heran = pd.read_csv('heran_data_csv.csv')
Hyosub = pd.read_csv('hyosub_data_csv.csv')
John = pd.read_csv('johnb_data_csv.csv')
Jon = pd.read_csv('jonwood_data_csv.csv')
Joie = pd.read_csv('joie_data_csv.csv')
Riwa = pd.read_csv('riwa_data_csv.csv')

#relabel dataframe columns:

Heran = Heran.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})
Hyosub = Hyosub.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})
John = John.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})
Jon = Jon.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})
Joie = Joie.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})
Riwa = Riwa.rename(columns = {'hand_angle_diff': 'Hand Angle Difference(degrees)'})

Heran.head(5)
```

# Plots of Subject Data

```python
#Get a list of targets
targets = np.unique(Heran['target_pos'])

#Create a list of colors to match to each target reach
c = [cm.hsv(x/len(targets)) for x in range(len(targets))]

#Create groupby object (pull out data from data frame, grouped by each target)
target_groups = Heran.groupby('target_pos')

#Plot 'reaches' for each subject, and color it relative to target location
x = Heran['xcoord']
y = Heran['ycoord']
fig, ax = plt.subplots(1,2, figsize = (10,5) )
ax[0].set_title("Heran's Reach Endpoints")
ax[1].set_title("Heran's Reach Error")

#Plot each target reach direction
for i in range(len(targets)):
    target = target_groups.get_group(targets[i])
    x = target['xcoord']
    y = target['ycoord']
    ax[0].scatter(x,y, color = c[i], label = str(targets[i])+ '°' )

#Plot legend for the reaches
ax[0].legend()

#Plot the histogram setting the range to exlude outliers
ax[1].hist(Heran['Hand Angle Difference(degrees)'], bins = 30, range = (-60,60), edgecolor='k', color = 'r')
ax[1].set_xlabel('Reach Error')
ax[1].set_ylabel('Frequency')
```
